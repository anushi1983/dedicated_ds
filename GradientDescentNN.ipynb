```python
def log_loss(yt,yp):
  eps=1e-15
  yp_new=[max(i,eps) for i in yp]
  yp_new=[min(i,1-eps) for i in yp_new]
  yp_new= np.array(yp_new)
  prec=-np.mean(yt*np.log(yp_new)+(1-yt)*np.log(1-yp_new))
  return prec
```

```python
def sigmoid_np(x):
  return 1/(1+np.exp(-x))
sigmoid_np(np.array([12,0,1]))
```

```python
class MyNN:
  def _init_(self):
    self.w1=1
    self.w2=1
    self.bias=0
  def fit(self,X,y,epochs,loss_threshold= 0.3426):
    self.w1,self.w2,self.bias=self.gradient_descent(X['age'],X['affordability'],y,epochs,loss_threshold)
  def predict(self,X):
    weighted_sum=self.w1*X['age']+self.w2*X['affordability']+self.bias
    predicted= self.sigmoid_np(weighted_sum)
    return(predicted)
  def sigmoid_np(self,x):
    return 1/(1+np.exp(-x)) 
  def gradient_descent(self,age,aff,yt,epochs,loss_threshold):
    #weighted sum and bias
    w1=w2=1
    bias=0.1
    rate=0.5
    n=len(age)
    for i in range(epochs):
      weighted_sum=w1*age+w2*aff+bias
      y= self.sigmoid_np(weighted_sum)
      dw=(1/n)*np.dot(np.transpose(age),(y-yt))
      db=(1/n)*np.dot(np.transpose(aff),(y-yt))
      bias_d=np.mean(y-yt)
      loss=log_loss(yt,y)
      w1=w1-(rate *dw)
      w2=w2-(rate *db)
      bias=bias-rate*bias_d
      if i%10==0:
        print(f'Epoch={i}, w1={w1}, w2:{w2}, bias:{bias},  loss:{loss}')
      if loss < loss_threshold :
        break
    return w1,w2,bias
```
